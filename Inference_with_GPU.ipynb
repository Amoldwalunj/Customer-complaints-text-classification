```python
index_to_label= {0: 'Autonomy, Equality, and Empowerment',
 1: 'Cancer',
 2: 'Communicable Diseases',
 3: 'Education and Vocational Skills',
 4: 'Food and Adequate Nutrition',
 5: 'Free, Affordable or Insured Healthcare',
 6: 'Fully-functional and well-equipped Health Facilities',
 7: 'Health Workers',
 8: 'High Quality, Inclusive and Respectful Health Services',
 9: 'Interpersonal Relationships and Social Support',
 10: 'Maternal, Newborn, Child Health',
 11: 'Menstrual & Menopausal Cycles',
 12: 'Mental Health',
 13: 'No Harmful Practices and GBV',
 14: 'Non-Communicable Diseases',
 15: 'Physical Activity and Rest',
 16: 'Policy and Social Welfare',
 17: 'Safety',
 18: 'Sexual and Reproductive Health',
 19: 'Sustainable Energy & Agriculture',
 20: 'Transportation and Road Safety',
 21: 'Water, Sanitation & Hygiene',
 22: 'Work & Financial Support'}


# Create a dictionary mapping from index to class name
#index_to_label = {i: label for i, label in enumerate(mlb.classes_)}

def labels_from_predictions(predictions):
    """Convert predictions to label names using the index_to_label dictionary."""
    return [[index_to_label[idx] for idx, val in enumerate(row) if val == 1] for row in predictions]

def predict_single(text):
    """Predict the labels for a single text input."""
    predictions, _ = model.predict([text])
    return labels_from_predictions(predictions)[0]

def predict_batch(texts):
    """Predict the labels for a batch of text inputs."""
    predictions, _ = model.predict(texts)
    return labels_from_predictions(predictions)




whwb_to_data_exchange_mapping = {'Autonomy, Equality, and Empowerment': 'Equality & Rights',
'Cancer': 'Health',
'Communicable Diseases': 'Health',
 'Education and Vocational Skills':'Education',
'Food and Adequate Nutrition': 'Health',
'Free, Affordable or Insured Healthcare': 'Health',
'Fully-functional and well-equipped Health Facilities':'Health',
'Health Workers':'Health',
'High Quality, Inclusive and Respectful Health Services':'Health',
'Interpersonal Relationships and Social Support':'Mental, Emotional & Social Wellbeing',
'Maternal, Newborn, Child Health':'Health',
'Menstrual & Menopausal Cycles':'Health',
'Mental Health':'Mental, Emotional & Social Wellbeing',
'No Harmful Practices and GBV':'Safety & Security',
'Non-Communicable Diseases':'Health',
'Physical Activity and Rest':'Health',
'Policy and Social Welfare':'Equality & Rights',
'Safety':'Safety & Security',
'Sexual and Reproductive Health':'Health',
'Sustainable Energy & Agriculture':'Environment',
'Transportation and Road Safety':'Environment',
'Water, Sanitation & Hygiene':'Environment',
'Work & Financial Support':'Economy'}

def convert_to_data_exchange_codebook(demands):
    already_seen = set()
    mapped = []
    for d in demands:
        converted = whwb_to_data_exchange_mapping[d]
        if converted not in already_seen:
            mapped.append(converted)
            already_seen.add(converted)
    return mapped

```

```python
import pandas as pd

df= pd.read_csv("final_categories.csv")
```

```python
import pandas as pd
import ast

# Assuming df is your DataFrame already loaded

# Function to safely evaluate strings to lists or return plain strings
def evaluate_string(value):
    try:
        return ast.literal_eval(value)
    except (ValueError, SyntaxError):
        return value

# Function to apply mapping to each category
def apply_mapping(value, mapping):
    # If the value is a list, apply mapping to each element
    if isinstance(value, list):
        return [mapping.get(item, item) for item in value]  # Use .get() to safely handle unmapped values
    else:
        # If it's a single item, apply mapping directly
        return mapping.get(value, value)

# Step 1: Evaluate string representations to lists or leave plain strings as is
df['final_category'] = df['final_category'].apply(evaluate_string)

# Step 2: Apply the mapping to each category
df['transformed_category'] = df['final_category'].apply(lambda x: apply_mapping(x, whwb_to_data_exchange_mapping))

# Display the DataFrame to verify the transformation
print(df[['final_category', 'transformed_category']].head())

```

```python
# Function to remove duplicates while preserving order
def remove_duplicates_preserve_order(value):
    seen = set()
    seen_add = seen.add
    return [x for x in value if not (x in seen or seen_add(x))]

# Updated function to join categories with '/' after removing duplicates
def join_categories(value):
    if isinstance(value, list):
        # Remove duplicates while preserving order
        unique_categories = remove_duplicates_preserve_order(value)
        # Join the unique list elements into a single string separated by '/'
        return '/'.join(unique_categories)
    else:
        # If it's already a single category string, return as is
        return value

# Apply the updated function to the 'transformed_category' column
df['transformed_category'] = df['transformed_category'].apply(join_categories)

# Display the DataFrame to verify the transformation
print(df[['final_category', 'transformed_category']].head())
```
